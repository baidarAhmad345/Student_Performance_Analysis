# EDA

# Gender Distribution

```python
plt.figure(figsize=(7,5))
ax=sns.countplot(x="Gender" , data=df , hue="Gender")
for container in ax.containers:
    ax.bar_label(container)
plt.show()
```

# Note: From the above, we can conclude that the number of females is greater than the number of males.

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Sample DataFrame setup for demonstration purposes
# df = pd.read_csv('your_data.csv')

# Group by 'LunchType' and calculate mean for each subject
gb2 = df.groupby("Gender").agg({
    "MathScore": "mean",
    "ReadingScore": "mean",
    "WritingScore": "mean"
}).reset_index()

# Melt the DataFrame to long format
gb2_melted = gb2.melt(id_vars="Gender", var_name="Subject", value_name="AverageScore")

# Pivot the data for heatmap
heatmap_data = gb2_melted.pivot(index="Gender", columns="Subject", values="AverageScore")

# Plot heatmap
plt.figure(figsize=(10, 6))
ax = sns.heatmap(heatmap_data, annot=True, cmap="YlGnBu", fmt=".2f")
plt.title("Average Scores  by Gender")
plt.xlabel("Subject")
plt.ylabel("Gender")

plt.show()
```

# Note: From the graph, we can observe that males have higher scores in Math, but they score lower in Reading and Writing.

# Distribution of Ethnic Group

```python
groupA=df.loc[(df["EthnicGroup"]=="group A")].count()
groupB=df.loc[(df["EthnicGroup"]=="group B")].count()
groupC=df.loc[(df["EthnicGroup"]=="group C")].count()
groupD=df.loc[(df["EthnicGroup"]=="group D")].count()
groupE=df.loc[(df["EthnicGroup"]=="group E")].count()
l=["group A" , "group B"  , "group C" , "group D" , "group E"]
list=[groupA["EthnicGroup"] , groupB["EthnicGroup"]  , groupC["EthnicGroup"] , groupD["EthnicGroup"] , groupE["EthnicGroup"]]
plt.pie(list , autopct='%1.1f%%' , labels=l)
plt.title('Distribution of Ethnic Groups')

plt.show()

```

# Lunch Type

```python
gb2 = df.groupby("LunchType").agg({
    "MathScore": "mean",
    "ReadingScore": "mean",
    "WritingScore": "mean"
}).reset_index()

# Melt the DataFrame to long format
gb2_melted = gb2.melt(id_vars="LunchType", var_name="Subject", value_name="AverageScore")

# Plot using seaborn
plt.figure(figsize=(8, 6))
ax = sns.barplot(x="LunchType", y="AverageScore", hue="Subject", data=gb2_melted)

# Add labels to the bars
for container in ax.containers:
    ax.bar_label(container, fmt='%.2f', label_type='edge')

plt.title("Average Scores by Lunch Type")
plt.xlabel("Lunch Type")
plt.ylabel("Average Score")
plt.legend(title="Subject")
plt.legend(title="Subject", bbox_to_anchor=(1.05, 1), loc='upper left')

plt.show()
```
